### Basic Plotting using Bokeh

```python
# import necessary libraries
from bokeh.plotting import figure
from bokeh.io       import output_file, show, output_notebook, push_notebook
from bokeh.models   import ColumnDataSource, HoverTool, CategoricalColorMapper
from bokeh.layouts  import row, column, gridplot
from bokeh.models   import Tabs, TabPanel
output_notebook()
```

```python
f = figure()
f.line([10,20,30,40,50],[60,70,80,40,50], line_width=4)
show(f)
```

### Plotting using Glyph

```python
# circle glyph

p = figure(x_axis_label= "x-axis", y_axis_label= "y-axis", tools= "pan, box_zoom")
p.circle(x=[50,40,30], y=[10,20,30], size=25, color="yellow", alpha=0.7, line_color="red")
show(p)
```

```python
# square glyph

t = figure(width=500, height=500)
t.square([10,12,13,14,15],[16,17,12,14,15], size=20, color="blue", alpha=0.6)
show(t)
```

```python
# Multiple lines

y = figure(width=500, height=500)
y.multi_line([[10,20,30],[10,20,10]],[[30,40,40],[10,60,20]],color=["red","blue"], alpha=[0.7,0.7], line_width=4)
show(y)
```

```python
# wedge glyph

z = figure(width=500, height=500)
z.wedge(x=[1,2,3], y=[1,2,3], radius=0.2, start_angle=0.4, end_angle=5.2, color="green", alpha=0.7, direction="clock")
show(z)
```

```python
# annular wedge glyph

q = figure(width=500, height=500)
q.annular_wedge(x=[10,20,30], y=[10,20,30], inner_radius=0.5, outer_radius=0.75, start_angle=0.5, end_angle=5.2, color="magenta", alpha=0.7)
show(q)
```

```python
# combine circle and line on the same plot

x = [10,20,30,40,50]
y = [60,75,80,95,55]

p = figure(width=500, height=500)
p.line(x,y, line_width=2)
p.circle(x,y, fill_color="black", size=9)

show(p)
```

### Creating Plots/Charts

```python
# Horizontal Bar Plot

g1 = figure(title="Horizontal Bar Plot", height=400, width=400)

x = [1,2,3,4,5]
y = [9,8,7,6,5]

g1.hbar(x, right=y, height=0.4, color="rebeccapurple")
show(g1)
```

```python
# Vertical Bar Plot

g2 = figure(title="Vertical Bar Plot", width=400, height=400)

x = [1,2,3,4,5]
y = [9,8,7,6,5]

g2.vbar(x=x, top=y, width=0.5, color="rebeccapurple")
show(g2)
```

```python
# Scatter Plot

import random

g3 = figure(title="Scatter Plot", width=400, height=400)

x = [i for i in range(126)]
y = [random.random()+2 for i in range(126)]

g3.scatter(x,y, color="firebrick")
show(g3)
```

```python
# Area Plot

import numpy as np

x = [10,20,30,40,50]
y1 = [1,3,4,5,6]
y2 = [2,4,7,8,9]

g4 = figure(width=400, height=400)
g4.varea(x=x, y1=y1, y2=y2, fill_color="cornflowerblue")
show(g4)
```

```python
# Patch Plot

g5 = figure(width=400, height=400)
g5.patch([1,4,6,9],[2,3,7,8], color="navy")
show(g5)
```

### Mapping my University !

```python
from bokeh.io import show, output_notebook
from bokeh.models import GMapOptions, ColumnDataSource
from bokeh.plotting import gmap

# 1. Setup Map Options
# Coordinates for Thapar Institute of Engineering and Technology (TIET)
latitude  = 30.3529
longitude = 76.3614
map_type  = "hybrid"
zoom      = 16

google_map_options = GMapOptions(lat=latitude, lng=longitude, map_type=map_type, zoom=zoom)

title = "Thapar University Map - Patiala"

google_api_key = ""

google_map = gmap(google_api_key, google_map_options, title=title)

# 2. Setup Data Source
source1 = ColumnDataSource(data=dict(
    lat=[30.3529],
    lon=[76.3614]
))

x = "lon"
y = "lat"
size = 20
fill_color = "red"
fill_alpha = 0.8

# 3. Add the glyphs
google_map.square(x=x, y=y, size=size, fill_color=fill_color, fill_alpha=fill_alpha, source=source1)

show(google_map)
```

```python
from bokeh.io import show, output_notebook
from bokeh.plotting import figure
import math

# 1. Convert Lat/Lon to Web Mercator
def to_mercator(lat, lon):
    r_major = 6378137.0
    x = r_major * math.radians(lon)
    scale = x / lon
    y = 180.0 / math.pi * math.log(math.tan(math.pi / 4.0 + lat * (math.pi / 180.0) / 2.0)) * scale
    return x, y

# Thapar University (TIET) Coordinates
lat, lon = 30.3529, 76.3614
x_coords, y_coords = to_mercator(lat, lon)

# 2. Setup Figure
# We define the range around Thapar (approx 1000 meters in each direction)
p = figure(x_range=(x_coords-1000, x_coords+1000),
           y_range=(y_coords-1000, y_coords+1000),
           x_axis_type="mercator", y_axis_type="mercator",
           title="Thapar University Map (TIET) - Bokeh 3.x",
           width=800,
           height=600)

# NEW WAY: Just pass the name string directly to add_tile
# Common names: "CartoDB Positron", "OSM", "Esri World Imagery"
p.add_tile("OSM")

# 3. Add the glyphs (Red Square for TIET)
p.square(x=[x_coords], y=[y_coords], size=20, fill_color="red", fill_alpha=0.8)

output_notebook()
show(p)
```
